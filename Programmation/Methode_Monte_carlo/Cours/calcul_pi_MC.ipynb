# Calcul de la valeur de Pi par la methode de Monte Carlo

Considerons le quart de cercle de rayon $1$ correspondant au premier quadrant, $0\le x;y \le 1$. On cherche à calculer sa superficie

on sait que: $$
            \begin{align*}
            x^2+y^2 =& 1  \text{ correspondant à l'equation du cercle de centre $(0;0)$ et de rayon $1$ }\\
            \text{ On} & \text{ endeduit l'expression de y en fonction de x}\\
            y=& \sqrt{1-x^2} \text{   }\forall x \in [0;1]
            \end{align*}
            $$

l'aire $A$ de ce quart de cercle represente l'aire sous la courbe de $f(x)=\sqrt{1-x^2}$ sur l'intervalle $[0;1]$:
$$
A = \int_{0}^{1} f(x) \,dx
$$
Nous allons alors estimer la valeur de cette integrale grace à la methode Monte Carlo.

1. Tirer aleatoirement $N$ valeurs $x_i$ sur $[0;1]$ suivant la loi uniforme.
2. On calcul pour chaque point $f(x_i)$
3. on fait ensuite la moyenne arithmétique $<f(x_i)>$ des $f(x_i)$. la valeur obtenu correspond à celle de l'aire A

On peut endeduire alors la valeur de $\pi$ grace à la formule de l'aire d'un disque: $A = \frac{\pi \times 1^2}{4}= \frac{\pi}{4}$

$\pi = 4 *A$
           

```python
import numpy as np 
def f(x):
    return np.sqrt(1-x**2)

def Integration_MC(a:float, b:float, N: float, f )-> float:
    """Estimer la valeur d'une integrale en dimension 1
    
    Args
    ----
    a: borne inferieur de l'integrale
    b: borne superieur de l'integrale
    N: Nombre total de valeur à tirer au hasard suivant la loi uniforme
    f: fonctions à integrer 

    retourne
    --------
    l'approximation de pi
    """

    
    i = 0
    somme_fx_i = 0
    somme_fx_i2 = 0
    v = b-a
    while i < N:
        x_i = np.random.random()
        f_xi = f(x_i)
        somme_fx_i +=  f_xi
        somme_fx_i2 +=  f_xi*f_xi
        i += 1
    aire = somme_fx_i/float(N) # <f(x)>
    somme_fx_i2 /= float(N)                               #<f(x)**2>
    pi = 4*aire                                           # Estimation de pi
    var_fx_i = somme_fx_i2 - aire**2                      #  variane = <f(x)**2> - <f(x)>**2
    Integration_MC.erreur = np.sqrt(var_fx_i/float(N))
    return pi 

 #if __main__ == __name__ : 
for N in [1e3,1e4, 1e5]:
    print( f" Pour N = {N}\n Pi = {Integration_MC(0,1,N, f)} +/- {Integration_MC.erreur}")




    
```

 On remarque que  la precision de la  methode Monte Carlo augmente avec le nombre totale  de points tirer.
