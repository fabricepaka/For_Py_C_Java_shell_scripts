# Introduction à la programmation modulaire
* La programmation modulaire est une technique de programmation dans laquelle on organise notre programme en plusieurs sous programme connectés les uns aux autres. 

* Un fichier source python est un fichier dans lequel on redige un programme python. Ce fichier nous évite de repéter les instructions demander à l'interprèteur à python. le nom de ce fichier source est toujours suivi dsu suffixe `.py`

* Au fur et à mesure que le programme qu'on ecrit dans ce fichier augmente, il devient difficile de suivre toutes les fonctions et classes definies dans le fichier source.

* La programmation modulaire se présente comme un moyen robuste d'organiser notre programme lorsqu'il devient de plus en plus compliqué et dense.

* On peut donc subdiviser un programme très dense en plusieurs sous programme servant à resoudre des problèmes spécifiques et les importer les uns dans les autres.

* Tout programme (ou code) python est un module. On peut l'utiliser comme programme, on peut l'utiliser comme bibliothèque( module)

* Pour importer un module, il sufﬁt de faire appel à la directive `import` suivi du nom du ﬁchier (sans le .py).
    - par exemple: ```python
                    import name``` 







### Les librairies python
De la meme facon que python te permet d'organiser les fonctions et les classes dans des fichiers separés, python te permet aussi de grouper plusieurs modules en librairie (ou package).

Une librairie python est un dossier contenant des modules python et qui possède des caractéristiques particulières:
  - il contient un fichier source nommé `__init__.py`, nommé le fichier d'initialisation du package.
  - la présence de ce fichier indique à python que ce dossier est une librairie.il intialise le package et execute des codes lorqu'il est importé 
  - On importer une fonction d'un module spécifique dans notre librairie un module spécifique en faisant:

         ```python
         from nom_de_la_librairie.nom_du_module import nom_de_la_fonction_souhaité
         ```

  
*Notons que ce fichier `__init__.py` n'est plus obligatoire dans les versions python superieur à 3.3 pour qu'un dossier contenant des modules python soit un module.

Cependant bien que ce ne soit pas une exigence, conserver un fichier `__init.py__` reste utile.

        

*La programmation spaghetti est le contraire de la programmation modulaire. Dans ce style de codage, on y met toutes les fonctions et classes dans un seul et unique script.*

### le shebang
On peut exécuter un code Python sans avoir à le précéder de `python` exécutable. 
Pour ce faire, on doit :
1. rendre le fichier exécutable, grace a la commande `chmod`

`chmod +x chemin-du-fichier`

2. indiquer quel interpréteur doit l'exécuter, grace au shebang en premiere ligne du fichier.

Un shebang commence par `#!`, puis est suivi du chemin vers l’interpréteur :

`#!/usr/bin/python3` par exemple
